```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns 
np.random.seed(42)

#Group A old website, 20% conversion 

group_a= np.random.binomial(n=1,p=0.20, size=1000)

#Group b new website, 24% conversion
group_b = np.random.binomial(n=1, p=0.24, size=1000)
print("Group A conversion rate:", round(np.mean(group_a) * 100, 2), "%")
print("Group B conversion rate:", round(np.mean(group_b) * 100, 2), "%")
print("Difference:", round((np.mean(group_b) - np.mean(group_a)) * 100, 2), "%")
```

```python
t_stat, p_value = stats.ttest_ind(group_a, group_b)
print("P_value", round(p_value, 4))

if p_value < 0.05:
    print("SIGNIFICANT — launch the new website")

else:
    print("NOT SIGNIFICANT — don't launch yet")
```

We tested a new website design against the old one to see which performed better. The new design had a 5.2% higher conversion rate, and the p-value was 0.0053, which means there’s only a very small chance that this improvement happened randomly. Since the result is statistically significant, the company should consider launching the new design because, at a larger scale, it could lead to thousands of extra conversions.

```python
labels = ['Group A (Old)', 'Group B (New)']
rates = [np.mean(group_a) * 100, np.mean(group_b) * 100]
colors = ['#ff6b6b', '#51cf66']

plt.figure(figsize=(8,5))
sns.barplot(x=labels, y=rates, palette=colors)
plt.title('A/B Test Results — Conversion Rate Comparison')
plt.ylabel('Conversion Rate %')
plt.ylim(0, 35)

for i, rate in enumerate(rates):
    plt.text(i, rate + 0.5, f'{rate:.1f}%', 
             ha='center', fontweight='bold', fontsize=12)

plt.show()
```
